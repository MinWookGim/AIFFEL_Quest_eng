#프로젝트1

```python
class Square:
    def __init__(self):
        self.square = int(input('넓이를 구하고 싶은 사각형을 선택하세요.\n 1.직사각형 2.평행사변형 3.사다리꼴 : '))

    # 직사각형 메서드
    def rect(self):
        width, vertical = map(int, input('가로, 세로를 입력하세요. 예시 : 가로,세로\n >>>').split(','))
        area = width * vertical
        result = '직사각형의 넓이는 : ' + str(area)
        return result

    # 평행사변형 메서드
    def par(self):
        width, vertical = map(int, input('가로, 세로를 입력하세요. 예시 : 가로,세로\n >>>').split(','))
        area = width * vertical
        result = '평행사변형 넓이는 : ' + str(area)
        return result

    # 사다리꼴 메서드
    def trape(self):
        up_l, dw_l, he_l = map(int, input('윗변, 아랫변, 높이를 입력하세요. 예시 : 윗변,아랫변,높이\n >>>').split(','))
        area = (up_l+dw_l)*he_l/2
        result = '사다리꼴 넓이는 : ' + str(area)
        return result

    def question(self):
        if self.square == 1:
            print(self.rect())
        elif self.square == 2:
            print(self.par())
        elif self.square == 3:
            print(self.trape())
        else:
            print("1, 2, 3 중에 1개만 입력하세요. 다시 하려거든 객체를 다시 생성하세요.")
```

```python
a = Square();

a.question()
```

#프로젝트2

```python
class Kiosk:
    def __init__(self):
        # 1. 메뉴와 가격 리스트 정의
        self.menu = ['americano', 'latte', 'mocha', 'yuza_tea', 'green_tea', 'choco_latte']
        self.price = [2000, 3000, 3000, 2500, 2500, 3000]
        # 2. 주문 내역을 담을 빈 리스트 초기화
        self.order_menu = []
        self.order_price = []

    # 메뉴 출력 메서드
    def menu_print(self):
        for i in range(len(self.menu)):
            print(f"{i+1} {self.menu[i]} : {self.price[i]}원")

    # 주문 메서드
    def menu_select(self):
        self.order_menu = []
        self.order_price = []
        
        n = 0
        while n != -1:
            n = int(input("음료 번호를 입력하세요 (주문 완료는 -1) : "))
            
            if n == -1:
                break
                
            # 메뉴 범위 내 숫자를 입력했을 때
            if 1 <= n <= len(self.menu):
                t = 0
                while t != 1 and t != 2:
                    t = int(input("HOT 음료는 1을, ICE 음료는 2를 입력하세요 : "))
                    if t == 1:
                        temp = 'HOT'
                    elif t == 2:
                        temp = 'ICE'
                    else:
                        print("1과 2 중 하나를 입력하세요.\n")
                
                # 주문 내역에 추가
                ordered_item = f"{temp} {self.menu[n-1]}"
                self.order_menu.append(ordered_item)
                self.order_price.append(self.price[n-1])
                
                print(f"주문 음료 : {ordered_item} : {self.price[n-1]}원")
            else:
                print("없는 메뉴 번호입니다. 다시 입력해 주세요.")

    # 지불 메서드
    def pay(self):
        p = 0
        while p != 1 and p != 2:
            p = int(input("현금 결제는 1을, 카드 결제는 2를 입력하세요 : "))
            if p == 1:
                self.payment = "현금"  # self.payment 변수에 저장
                print("직원을 호출하겠습니다.")
            elif p == 2:
                self.payment = "카드"  # self.payment 변수에 저장
                print("IC칩 방향에 맞게 카드를 꽂아주세요.")
            else:
                print("다시 결제를 시도해 주세요.")

    # 주문서 출력 메서드
    def table(self):
        # 외곽선 그리기 (상단)
        print('⟝' + '-' * 30 + '⟞')
        for i in range(5):
            print('|' + ' ' * 31 + '|')
            
        # 주문 상품명 : 해당 금액 출력
        for i in range(len(self.order_menu)):
            print(self.order_menu[i], ' : ', self.order_price[i])
        
        # 💡 결제 수단 및 합계 금액 출력
        print('-' * 32)
        print('결제 수단 :', self.payment)  # 저장해둔 결제 수단 출력!
        print('합계 금액 :', sum(self.order_price))
        
        # 외곽선 그리기 (하단)
        for i in range(5):
            print('|' + ' ' * 31 + '|')
        print('⟝' + '-' * 30 + '⟞')
```

```python
a = Kiosk()  # 객체 생성 
a.menu_print()  # 메뉴 출력
a.menu_select()  # 주문
a.pay()  # 지불
a.table()  # 주문표 출력
```

# 프로젝트2 확장

```python
from datetime import datetime

# 데코레이터 정의 (주문표 상단/하단 양식 출력)
def receipt_decorator(func):
    def wrapper(self):
        print('⟝' + '-' * 30 + '⟞')
        for _ in range(2):  # 여백을 살짝 줄여서 깔끔하게 조정
            print('|' + ' ' * 31 + '|')
        
        # 원래의 table() 메서드 내용 실행
        func(self)
        
        for _ in range(2):
            print('|' + ' ' * 31 + '|')
        print('⟝' + '-' * 30 + '⟞')
    return wrapper

class Kiosk:
    def __init__(self):
        # 메뉴와 가격 리스트를 다차원 리스트로 변경
        self.menu_list = [
            ['americano', 2000],
            ['latte', 3000],
            ['mocha', 3000],
            ['yuza_tea', 2500],
            ['green_tea', 2500],
            ['choco_latte', 3000]
        ]
        self.order_menu = []
        self.order_price = []
        self.payment = ""

    # 메뉴 출력 시 enumerate() : enumerate - 시퀀스시 인덱스와 값을 같이 반환해준다.
    def menu_print(self):
        for idx, (menu_name, menu_price) in enumerate(self.menu_list, start=1): 
            print(f"{idx} {menu_name} : {menu_price}원")

    # 주문
    def menu_select(self):
        self.order_menu = []
        self.order_price = []
        
        n = 0
        while n != -1:
            n = int(input("음료 번호를 입력하세요 (주문 완료는 -1) : "))
            
            if n == -1:
                break
                
            if 1 <= n <= len(self.menu_list):
                t = 0
                while t != 1 and t != 2:
                    t = int(input("HOT 음료는 1을, ICE 음료는 2를 입력하세요 : "))
                    if t == 1:
                        temp = 'HOT'
                    elif t == 2:
                        temp = 'ICE'
                    else:
                        print("1과 2 중 하나를 입력하세요.\n")
                
                # 2차원 리스트에서 이름과 가격 꺼내기
                menu_name = self.menu_list[n-1][0]
                menu_price = self.menu_list[n-1][1]
                
                ordered_item = f"{temp} {menu_name}"
                self.order_menu.append(ordered_item)
                self.order_price.append(menu_price)
                
                print(f"주문 음료 : {ordered_item} : {menu_price}원")
            else:
                print("없는 메뉴 번호입니다. 다시 입력해 주세요.")

    # 지불 메서드
    def pay(self):
        p = 0
        while p != 1 and p != 2:
            p = int(input("현금 결제는 1을, 카드 결제는 2를 입력하세요 : "))
            if p == 1:
                self.payment = "현금"
                print("직원을 호출하겠습니다.")
            elif p == 2:
                self.payment = "카드"
                print("IC칩 방향에 맞게 카드를 꽂아주세요.")
            else:
                print("다시 결제를 시도해 주세요.")

    # 데코레이터 적용 / datetime 주문 일시 추가
    @receipt_decorator
    def table(self):
        # 3. datetime 모듈로 현재 주문 일시 생성
        now = datetime.now().strftime('%Y-%m-%d %H:%M:%S') #형식 맞추기
        print(f"| 주문 일시 : {now}")
        print('|' + '-' * 30 + '|')
            
        # 주문 상품명 : 해당 금액 출력
        for i in range(len(self.order_menu)):
            print(f"| {self.order_menu[i]} : {self.order_price[i]}원")
            
        print('|' + '-' * 30 + '|')
        print(f"| 결제 수단 : {self.payment}")
        print(f"| 합계 금액 : {sum(self.order_price)}원")
```

```python
# 1. 고도화된 키오스크 객체 생성
cafe_kiosk = Kiosk()

# 2. 메뉴판 출력 (enumerate 활용 버전)
cafe_kiosk.menu_print()

# 3. 주문하기 (2차원 리스트 활용 버전)
cafe_kiosk.menu_select()

# 4. 결제하기 (결제 수단 저장 버전)
cafe_kiosk.pay()

# 5. 주문표 출력 (데코레이터 및 실시간 datetime 적용 버전)
cafe_kiosk.table()
```
