#Q1: 은행계좌 만들기

```python
import random # 계좌번호 랜덤

class Account:
    # [Q2] 클래스 변수: 모든 인스턴스가 공유하는 값. 계좌 개수를 여기에 누적
    account_count = 0

    # [Q1] 생성자(__init__) = 매직메서드. 예금주와 초기 잔액만 입력받음
    def __init__(self, owner, balance):
        self.bank_name = "SC은행" # 고정값
        self.owner = owner
        self.balance = balance

        # [Q1] 계좌번호 3자리-2자리-6자리 랜덤 생성
        num1 = random.randint(0, 999)
        num2 = random.randint(0, 99)
        num3 = random.randint(0, 999999)
        self.account_number = f"{num1:03d}-{num2:02d}-{num3:06d}"

        # [Q2] 인스턴스가 만들어질 때마다 클래스 변수 +1
        Account.account_count += 1

        # [Q7] 이자 계산용 입금 횟수
        self.deposit_count = 0
        # [Q10] 입출금 내역 저장용 리스트
        self.deposit_log = []
        self.withdraw_log = []

    # [Q3] 클래스 변수 출력
    @classmethod  # @classmethod는 파이썬 메모리 heap영역 중 클래스의 레퍼런스를 제공해주는 데코레이터이다. 즉, Account 클래스의 레퍼런스를 제공해준다. (매개변수 cls)
    def get_account_num(cls):
        print(f"총 계좌 개수: {cls.account_count}")

    # [Q4] 입금: 최소 1원 이상
    def deposit(self, amount):
        if amount < 1:
            print("입금은 최소 1원 이상만 가능합니다.")
            return
        self.balance += amount
        self.deposit_count += 1
        self.deposit_log.append(amount)        # [Q10]

        # [Q7] 입금 5회마다 잔고의 1% 이자 지급
        if self.deposit_count % 5 == 0:
            interest = self.balance * 0.01
            self.balance += interest
            print(f"이자 {interest:,.0f}원이 지급되었습니다.")

    # [Q5] 출금: 잔고보다 많이 못 뺌 (잔고보다 많이 출금하면 은행 망하니까.)
    def withdraw(self, amount):
        if amount > self.balance:
            print("잔고 이상으로 출금할 수 없습니다.")
            return
        self.balance -= amount
        self.withdraw_log.append(amount)       # [Q10]

    # [Q6] 정보 출력: 잔고는 천 단위 콤마
    def display_info(self):
        print(f"은행이름: {self.bank_name}, 예금주: {self.owner}, "
              f"계좌번호: {self.account_number}, 잔고: {self.balance:,.0f}원")

    # [Q10] 입금 내역 / 출금 내역
    def deposit_history(self):
        print(f"[{self.owner}님 입금 내역]")
        for amt in self.deposit_log:
            print(f"  +{amt:,}원")

    def withdraw_history(self):
        print(f"[{self.owner}님 출금 내역]")
        for amt in self.withdraw_log:
            print(f"  -{amt:,}원")
```

```python
# [Q8] 인스턴스 3개 이상 + 리스트에 저장
accounts = [
    Account("김철수", 500_000),
    Account("이영희", 1_500_000),
    Account("박민수", 2_000_000),
]

# [Q9] 잔고 100만원 이상 고객만 출력
for acc in accounts:
    if acc.balance >= 1_000_000:
        acc.display_info()

# [Q10] 입출금 내역 확인
accounts[1].deposit(300_000)
accounts[1].withdraw(100_000)
accounts[1].deposit_history()
accounts[1].withdraw_history()

Account.get_account_num()   # [Q3] → 총 계좌 개수: 3
```

# Q2: 간단한 자동사냥 RPG 만들기

```python
import random


# Q1. Character 클래스 만들기
class Character:

    def __init__(self, name, level, hp, attack, defense):
        self.name = name          # 이름
        self.level = level        # 레벨
        self.hp = hp              # 체력
        self.attack = attack      # 공격력
        self.defense = defense    # 방어력

    # 살아있는지 확인하는 is_alive 매서드 만들기
    def is_alive(self):
        return self.hp > 0

    # 공격 받았을 때 체력 주는 take_damage 매서드 만들기
    def take_damage(self, damage):
        real_damage = damage - self.defense

        if real_damage < 0:
            real_damage = 0

        self.hp -= real_damage

        if self.hp < 0:
            self.hp = 0

        print(f"{self.name}의 체력 : {self.hp}")

    # 공격하는 attack_target 매서드 만들
    def attack_target(self, target):
        damage = random.randint(1, self.attack)

        print(f"{self.name}이(가) {target.name}에게 {damage} 만큼 공격했다...!")

        target.take_damage(damage)



# Q2. Player 클래스 만들기
class Player(Character):

    def __init__(self, name):
        super().__init__(name, 1, 100, 25, 5)
        self.experience = 0  #경험치 속성 추가로 가짐

    # 경험치 획득하는 gain_experience 매서드 만들기
    def gain_experience(self, exp):
        self.experience += exp
        print(f"{exp} 경험치를 얻었습니다.")

    # 레벨업시키는 level_up 매서드 만들기
    def level_up(self):
        if self.experience >= 50:
            self.level += 1
            self.attack += 10
            self.defense += 5
            self.experience -= 50

            print("레벨업!")
            print(f"현재 레벨 : {self.level}")
            print(f"공격력 : {self.attack}")
            print(f"방어력 : {self.defense}")



# Q2. Monster 클래스 만들기
class Monster(Character):

    def __init__(self, name, level):
        hp = random.randint(10, 30) * level
        attack = random.randint(5, 20) * level
        defense = random.randint(1, 5) * level

        super().__init__(name, level, hp, attack, defense)



# Q3. battle 함수 만들기
def battle(player, monster):

    print(f"{monster.name} 과의 전투를 시작합니다.")

    while player.is_alive() and monster.is_alive():

        # 플레이어가 몬스터 공격
        player.attack_target(monster)

        # 몬스터가 죽으면 반복 종료
        if not monster.is_alive():
            break

        # 몬스터가 플레이어 공격
        monster.attack_target(player)

    # 전투 결과
    if player.is_alive():
        print("전투 승리!")

        player.gain_experience(monster.level * 20)
        player.level_up()

        print(f"{monster.name}을 이겼다!")

    else:
        print("전투 패배..")



# Q4. main 함수 만들기
def main():

    monster_dict = {"슬라임": 1,"고블린": 2,"오크": 3}

    player_name = input("플레이어 이름을 입력하세요 : ")

    player = Player(player_name)

    for monster_name, monster_level in monster_dict.items():

        monster = Monster(monster_name, monster_level)

        battle(player, monster)

        if not player.is_alive():
            print("게임오버")
            return

    print("모든 몬스터를 물리쳤습니다.!")
    print("게임 클리어!")



# 프로그램 실행
main()
```
