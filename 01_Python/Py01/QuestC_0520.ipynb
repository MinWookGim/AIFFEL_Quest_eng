#[Q1-1] 회문 판독

##회문: 앞에서부터 읽으나 뒤에서부터 읽으나 같은 말이 되는 단어, 구, 문장, 숫자 혹은 문단.

```python
def is_hoimun(strdata):
    return all( strdata[i] == strdata[-(i+1)] for i in range(len(strdata) //2))

print(is_hoimun("아많다많다많다많아"))  #True
print(is_hoimun("아참말많다많다많아"))  #False
```

#[Q1-2] 회문 분류기

##[Q1-1]에서 만든 판독기 함수를 이용해 회문 분류기 함수를 만든다.

```python
def is_hoimun_cls(strdata):
    result_dic = {
         '회문O': []
        ,'회문X': []
    }

    for word in strdata:
        if is_hoimun(word):
            result_dic['회문O'].append(word)
        else:
            result_dic['회문X'].append(word)

    tot_hoimun = len(result_dic['회문O'])
    tot_words = len(strdata)
    hoimun_ratio = tot_hoimun / tot_words if tot_words > 0 else 0.0

    return result_dic, tot_hoimun, hoimun_ratio


strdata = ['기러기', '까마귀', '토마토', '스윗소스', '스위스', '인도인', '한국인', '별똥별', '무영우', '무향우']
print(is_hoimun_cls(strdata))
```

#[Q2-1] 성적데이터화

```python
# 학생들의 성적 데이터를 딕셔너리 형태로 저장 [cite: 45, 46, 47]
scores = {
    "유원영": {"국어": 94, "영어": 91, "수학": 89},
    "설지민": {"국어": 88, "영어": 96, "수학": 90},
    "신윤아": {"국어": 86, "영어": 88, "수학": 77},
    "장윤아": {"국어": 98, "영어": 76, "수학": 92}
}
```

##[Q2-2] 성적분석기

```python
def score_analysis(name):
    if name not in scores:
        print("해당 학생의 데이터가 없습니다.")
        return
    
    # 학급 전체 평균
    num_students = len(scores)
    total_kor = sum(student["국어"] for student in scores.values())
    total_eng = sum(student["영어"] for student in scores.values())
    total_math = sum(student["수학"] for student in scores.values())
    
    avg_kor = total_kor / num_students
    avg_eng = total_eng / num_students
    avg_math = total_math / num_students
    
    # 학급 전과목 평균
    class_total_avg = (total_kor + total_eng + total_math) / (num_students * 3)
    
    # 해당 학생 성적 추출, 전과목 평균
    st_kor = scores[name]["국어"]
    st_eng = scores[name]["영어"]
    st_math = scores[name]["수학"]
    st_total_avg = (st_kor + st_eng + st_math) / 3
    
    # 평균 대비 차이
    diff_kor = st_kor - avg_kor
    diff_eng = st_eng - avg_eng
    diff_math = st_math - avg_math
    diff_total = st_total_avg - class_total_avg
    
    # 부호를 포함 소수점 둘째 자리 포매팅
    def format_diff(diff):
        return f"+{diff:.2f}" if diff > 0 else f"{diff:.2f}"
    
    # 결과 출력 포맷팅
    print(f"[{name} 성적 분석]")
    print(f"국어: {st_kor}점, 학급 평균({avg_kor:.2f}점) 대비 {format_diff(diff_kor)}")
    print(f"영어: {st_eng}점, 학급 평균({avg_eng:.2f}점) 대비 {format_diff(diff_eng)}")
    print(f"수학: {st_math}점, 학급 평균({avg_math:.2f}점) 대비 {format_diff(diff_math)}")
    print(f"전과목 평균: {st_total_avg:.2f}점, 학급 평균({class_total_avg:.2f}점) 대비 {format_diff(diff_total)}")

# 테스트
score_analysis("설지민")
score_analysis("장윤아")
score_analysis("신윤아")
score_analysis("유원영")
```
